## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.

### Cleaning Data

```python
# Konversi kolom tanggal
kolom_tanggal = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for kol in kolom_tanggal:
    df_orders[kol] = pd.to_datetime(df_orders[kol], errors='coerce')

print('Tipe data setelah konversi:')
print(df_orders[kolom_tanggal].dtypes)
```

```python
# Tangani missing values pada products
print('Missing sebelum:')
print(df_products[['product_category_name', 'product_weight_g']].isnull().sum())

df_products['product_category_name'].fillna(
    df_products['product_category_name'].mode()[0], inplace=True
)
df_products['product_weight_g'].fillna(
    df_products['product_weight_g'].median(), inplace=True
)

print('Missing sesudah:')
print(df_products[['product_category_name', 'product_weight_g']].isnull().sum())
```

```python
# Gabungkan dataset menjadi satu tabel analisis
df_products_merged = df_products.merge(df_category, on='product_category_name', how='left')

df_items_merged = df_order_items.merge(
    df_products_merged[['product_id', 'product_category_name_english']],
    on='product_id', how='left'
)

df_main = df_orders.merge(df_items_merged, on='order_id', how='inner')
df_main = df_main.merge(df_reviews[['order_id', 'review_score']], on='order_id', how='left')
df_main = df_main.merge(
    df_customers[['customer_id', 'customer_state']], on='customer_id', how='left'
)
df_main = df_main.merge(
    df_payments[['order_id', 'payment_type', 'payment_value']], on='order_id', how='left'
)

df_main['revenue'] = df_main['price'] + df_main['freight_value']
df_main['order_month'] = df_main['order_purchase_timestamp'].dt.to_period('M')
df_main['order_month_str'] = df_main['order_purchase_timestamp'].dt.strftime('%Y-%m')

print('Shape tabel gabungan:', df_main.shape)
df_main.head(3)
```

## Analisis Lanjutan: RFM Analysis
RFM Analysis digunakan untuk mengelompokkan pelanggan berdasarkan tiga aspek perilaku belanja:
- **Recency**: berapa hari sejak transaksi terakhir
- **Frequency**: berapa kali pelanggan melakukan pembelian
- **Monetary**: total nilai belanja pelanggan

```python
tgl_ref = df_delivered['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

df_rfm_raw = df_delivered.merge(
    df_customers[['customer_id', 'customer_unique_id']], on='customer_id', how='left'
)

df_rfm = (
    df_rfm_raw
    .groupby('customer_unique_id')
    .agg(
        Recency=('order_purchase_timestamp', lambda x: (tgl_ref - x.max()).days),
        Frequency=('order_id', 'nunique'),
        Monetary=('revenue', 'sum')
    )
    .reset_index()
)

print(df_rfm.describe())
```

```python
df_rfm['R_score'] = pd.qcut(df_rfm['Recency'],  q=5, labels=[5,4,3,2,1], duplicates='drop').astype(int)
df_rfm['F_score'] = pd.qcut(df_rfm['Frequency'].rank(method='first'), q=5, labels=[1,2,3,4,5], duplicates='drop').astype(int)
df_rfm['M_score'] = pd.qcut(df_rfm['Monetary'],  q=5, labels=[1,2,3,4,5], duplicates='drop').astype(int)
df_rfm['RFM_score'] = df_rfm['R_score'] + df_rfm['F_score'] + df_rfm['M_score']

def segmen(row):
    r, f, m = row['R_score'], row['F_score'], row['M_score']
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif r >= 4 and f >= 3:
        return 'Loyal Customers'
    elif r >= 3 and f >= 3 and m >= 3:
        return 'Potential Loyalist'
    elif r <= 2 and f >= 3:
        return 'At Risk'
    elif r <= 2 and f <= 2 and m <= 2:
        return 'Lost'
    else:
        return 'New Customers'

df_rfm['segment'] = df_rfm.apply(segmen, axis=1)
print(df_rfm['segment'].value_counts())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

seg_count = df_rfm['segment'].value_counts()
axes[0].bar(seg_count.index, seg_count.values, color='steelblue')
axes[0].set_title('Jumlah Pelanggan per Segmen RFM')
axes[0].set_ylabel('Jumlah Pelanggan')
axes[0].tick_params(axis='x', rotation=30)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

avg_m = df_rfm.groupby('segment')['Monetary'].mean().sort_values(ascending=True)
axes[1].barh(avg_m.index, avg_m.values, color='steelblue')
axes[1].set_title('Rata-rata Total Belanja per Segmen')
axes[1].set_xlabel('Rata-rata Belanja (R$)')
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

**Insight:** (Opsional)
- Total 93.358 pelanggan unik berhasil dianalisis menggunakan RFM
- Mayoritas pelanggan (rata-rata Frequency = 1.03) hanya melakukan 1 kali pembelian, mengindikasikan rendahnya tingkat repeat purchase di platform ini
- Segmen terbesar adalah New Customers (35.611) diikuti At Risk (22.230), yang berarti banyak pelanggan baru namun tidak kembali berbelanja
- Champions hanya berjumlah 6.463 pelanggan (6.9% dari total), namun memiliki rata-rata belanja tertinggi (R$ 320+) berdasarkan grafik kanan
- Segmen Lost (6.260) perlu diperhatikan karena mereka adalah pelanggan yang sudah tidak aktif sama sekali

Geospatial Analysis bertujuan menganalisis data berdasarkan lokasi geografis untuk mengidentifikasi tren atau pola tertentu di suatu wilayah.

```python
geo_stats = (
    df_delivered
    .groupby('customer_state')
    .agg(
        total_orders=('order_id', 'nunique'),
        total_revenue=('revenue', 'sum')
    )
    .reset_index()
    .sort_values('total_orders', ascending=False)
)

print(geo_stats.head(10).to_string(index=False))
```

```python
coords = {
    'SP': (-23.5, -46.6), 'RJ': (-22.9, -43.2), 'MG': (-19.9, -43.9),
    'RS': (-30.0, -51.2), 'PR': (-25.4, -49.3), 'SC': (-27.6, -48.5),
    'BA': (-12.9, -38.5), 'GO': (-16.7, -49.3), 'PE': (-8.1,  -34.9),
    'CE': (-3.7,  -38.5), 'ES': (-19.2, -40.3), 'MT': (-12.6, -55.9),
    'MS': (-20.4, -54.6), 'PB': (-7.2,  -36.8), 'RN': (-5.8,  -36.5),
    'AL': (-9.7,  -36.6), 'MA': (-5.4,  -45.4), 'PI': (-8.0,  -43.0),
    'PA': (-3.4,  -52.0), 'AM': (-4.4,  -63.6), 'DF': (-15.8, -47.9)
}

geo_stats['lat'] = geo_stats['customer_state'].map(lambda x: coords.get(x, (-15, -50))[0])
geo_stats['lon'] = geo_stats['customer_state'].map(lambda x: coords.get(x, (-15, -50))[1])

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# bubble map jumlah pesanan
ukuran = geo_stats['total_orders'] / geo_stats['total_orders'].max() * 1500
axes[0].scatter(geo_stats['lon'], geo_stats['lat'], s=ukuran,
                alpha=0.6, edgecolors='gray', linewidth=0.5)
for _, row in geo_stats.iterrows():
    axes[0].text(row['lon'], row['lat'], row['customer_state'],
                 ha='center', va='center', fontsize=7)
axes[0].set_title('Distribusi Jumlah Pesanan per Negara Bagian')
axes[0].set_xlabel('Longitude')
axes[0].set_ylabel('Latitude')
axes[0].set_xlim(-75, -30)
axes[0].set_ylim(-35, 5)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

# bar chart top 10
top10 = geo_stats.head(10)
axes[1].barh(top10['customer_state'], top10['total_orders'], color='steelblue')
axes[1].set_title('Top 10 Negara Bagian - Jumlah Pesanan')
axes[1].set_xlabel('Jumlah Pesanan')
axes[1].invert_yaxis()
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

- Rio de Janeiro (RJ) dan Minas Gerais (MG) berada di posisi 2 dan 3, namun jumlahnya hanya sekitar 30% dari SP
- Distribusi pesanan sangat tidak merata — mayoritas transaksi terkonsentrasi di wilayah tenggara Brasil (SP, RJ, MG), yang merupakan pusat ekonomi Brasil
